```python
import pandas as pd
from tqdm import tqdm
import os 
```

```python
test_accessions = open("../../../data/processed_data/genome_partitions/test_partition_accessions.txt").read().splitlines()
```

```python
error_models = ["complete"] #no sequencing errors
data_dirs = ["without_errors_300bp"] #MODIFY

for error_model in error_models:
    for data_dir in data_dirs:

        print(f"Predicting with error model: {error_model}")

        for test_accession in tqdm(test_accessions):
            #Run FragGeneScanRs
            path = f"../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}"
            os.makedirs(path, exist_ok=True)    

            gz_path = f"../../../data/processed_data/reads_processed/test/{data_dir}/fasta/{test_accession}.fasta.gz"
            if not os.path.isfile(gz_path):
                raise FileNotFoundError(f"GZIP file not found: {gz_path}")
            os.system(f"gunzip -k ../../../data/processed_data/reads_processed/test/{data_dir}/fasta/{test_accession}.fasta.gz")
            os.system(f"FragGeneScanRs -s ../../../data/processed_data/reads_processed/test/{data_dir}/fasta/{test_accession}.fasta -t {error_model} -w 0 -o ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}")

            #Clean up .gff output file 
            os.system(f"grep '	+	' ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}.gff > ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}+.gff")
            os.system(f"mv ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}+.gff ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}.gff")

            #Clean up .out output file
            os.system(f"grep -B 1 '	+	' ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}.out | grep -v '^--$' > ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}+.out")
            os.system(f"mv ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}+.out ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}.out")

            #Clean up .faa output file
            os.system(f"grep -A 1 '_+' ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}.faa | grep -v '^--$' > ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}+.faa")
            os.system(f"mv ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}+.faa ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}.faa")


            #Clean out .ffn output file
            os.system(f"grep -A 1 '_+' ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}.ffn | grep -v '^--$' > ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}+.ffn")
            os.system(f"mv ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}+.ffn ../../../data/processed_data/predictions/raw_predictions/fgs_preds/{data_dir}/{test_accession}/{test_accession}.ffn")

            os.system(f"rm ../../../data/processed_data/reads_processed/test/{data_dir}/fasta/{test_accession}.fasta")
```
